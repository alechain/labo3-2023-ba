```python
import glob

# Specify the directory pattern
directory_pattern = "/home/all4data118/buckets/b1/exp/model_es_b/product/model_trial*"

# Use glob to list directories matching the pattern
matching_directories = glob.glob(directory_pattern)
```

```python
import re

# Function to extract the model_trial value from the filename
def extract_trial_number(filename):
    match = re.search(r'model_trial_(\d+)', filename)
    return int(match.group(1)) if match else None

# Filter directories based on the model_trial value
filtered_directories = [
    directory for directory in matching_directories
    if 5 < extract_trial_number(directory) < 48
]

# Display the filtered directories
filtered_directories
```

```python
import re

# Function to extract the model_trial and rmse values from the filename
def extract_trial_and_rmse(filename):
    match = re.search(r'model_trial_(\d+)_seed_(\d+)_rmse_(\d+\.\d+)', filename)
    if match:
        trial_number, seed, rmse = map(float, match.groups())
        return trial_number, seed, rmse
    return None

# Filter directories based on the model_trial and rmse values
filtered_directories = [
    directory for directory in matching_directories
    if 5 < extract_trial_and_rmse(directory)[0] < 45
    and extract_trial_and_rmse(directory)[2] < 45  # Add this condition for rmse
]

# Display the filtered directories
filtered_directories
```

```python
from sklearn.preprocessing import StandardScaler
scalers = joblib.load('/home/all4data118/buckets/b1/datasets/scalers_product_tn_train_es_b.joblib')


# List to store the transformed DataFrames
transformed_dfs = []

# Iterate over each product_id in predictions_df
for product_id, scaler in scalers.items():
    # Select rows where product_id matches
    product_df = predictions_df[predictions_df['product_id'] == product_id].copy()

    # Check if the DataFrame is not empty
    if not product_df.empty:
        # Exclude 'product_id' from scaling
        cols_to_scale = product_df.columns.difference(['product_id'])

        # Inverse transform the selected columns
        product_df[cols_to_scale] = scaler.inverse_transform(product_df[cols_to_scale])

        # Append the transformed DataFrame to the list
        transformed_dfs.append(product_df)

# Concatenate the transformed DataFrames back together
transformed_predictions_df = pd.concat(transformed_dfs)

# Display or further process transformed_predictions_df
print(transformed_predictions_df)
```
